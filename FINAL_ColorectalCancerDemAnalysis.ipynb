```python
#import libraries
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import random as rnd

# read data into a DataFrame
data = pd.read_csv('/Users/kaushikshivakumar/Desktop/Stuff On Desktop/FilteredColoData.csv', index_col=0)
print data.shape
x = list(data.columns.values)
data
```

```python
#find which columns have the strongest correlations
thresh = 0.3
importantCols = []
valone = 0
valtwo = 1
count = 0
for i in range (3, 72):
    newFrame = data[['IncidencesPTTP', list(data.columns.values)[i]]]
    currentVal = (newFrame.corr('pearson')).iloc[0, 1]
    if (currentVal > thresh or currentVal < -thresh):
        importantCols.append(list(data.columns.values)[i])
        #importantCols.append(currentVal)
        count = count+1
print count
importantCols
```

```python
#filter a new dataframe down to these columns
importantData = data[['MentallyUnhealthyDays',
 'PercentObese',
 'PercentPhysicallyInactive',
 'PreventableHospRate',
 'PercentSomeCollege',
 'AssociationRate',
 'InjuryDeathRate',
 'AgeAdjustedMortality',
 'PercentDiabetic',
 'DrugOverdoseMortalityRate',
 'Medianhouseholdincome',
 'PercentLessThan18',
 'Percent65andover',
 'PercentHispanic',
 'PercentNonHispanicWhite',
 'PercentNotProficientinEnglish',
 'PercentRural', 'IncidencesPTTP']]
pd.options.display.max_columns = 100
pd.options.display.max_rows = 1000
importantData.shape
```

```python
#normalize the data
for i in range (0, 17):
    print i
    avg = 0;
    csum = 0;
    ccount = 0;
    cavg = 0;
    cboolarr = importantData.isnull()
    for j in range (0, 996):
        cval = importantData.iloc[j, i]
        cbool = cboolarr.iloc[j, i]
        if (not cbool):
            csum = csum + int(cval)
            ccount = ccount + 1
    cavg = csum/ccount
    for j in range (0, 996):
        if (not cbool):
            importantData.iloc[j, i] = (importantData.iloc[j, i])/cavg
```

```python
#create the model and assess statistical significance
lm = smf.ols(formula='IncidencesPTTP ~ MentallyUnhealthyDays + PercentObese + PercentPhysicallyInactive + PreventableHospRate + PercentSomeCollege + AssociationRate + InjuryDeathRate + AgeAdjustedMortality + PercentDiabetic + DrugOverdoseMortalityRate + Medianhouseholdincome + PercentLessThan18 + Percent65andover + PercentHispanic + PercentNonHispanicWhite + PercentNotProficientinEnglish + PercentRural', data = importantData).fit();
lm.summary()
```
